```python
# Imports and Magics required for plotting animations

%matplotlib nbagg 
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

# Setup the figure
fig = plt.figure()
fig.set_dpi(100)
fig.set_size_inches(8, 8)
ax = plt.axes(xlim=(300, 700), ylim=(300, 700))

# Create objects to be plotted
patch1 = plt.Circle((500, -500), 1, fc='r')
patch2 = plt.Circle((500, -500), 10, fc='r')

# Create and add into the figure, objects which will NOT be redrawn every frame
path1 = plt.Circle((500, 500), 0.0001, fill=False, fc='k')
path2 = plt.Circle((500, 500), 100, fill=False, fc='k')
vertLine = plt.Line2D((500, 500), (499.8, 500.2), c='g')
horLine = plt.Line2D((499.8, 500.2), (500, 500), c='g')
ax.add_patch(path1)
ax.add_patch(path2)
ax.add_artist(vertLine)
ax.add_artist(horLine)

# Initialisation function
def init():
    patch1.center = (5, 5)
    patch2.center = (5, 5)
    ax.add_patch(patch1)
    ax.add_patch(patch2)
    return patch1, patch2, 

# Animator function
def animate(i):
    x1, y1 = patch1.center
    x2, y2 = patch2.center
    x1 = 500 + 100 * np.sin(np.radians(i))
    y1 = 500 + 100 * np.cos(np.radians(i))
    x2 = 500 + 1e-4 * np.sin(np.radians(i) - np.pi)
    y2 = 500 + 1e-4 * np.cos(np.radians(i) - np.pi)
    patch1.center = (x1, y1)
    patch2.center = (x2, y2)
    return patch1, patch2, 

# Call the animator
anim = animation.FuncAnimation(fig, animate, 
                               init_func=init, 
                               frames=360, 
                               interval=100,
                               blit=True)

# Show the animation
#ax.axis(False)
plt.show()
```

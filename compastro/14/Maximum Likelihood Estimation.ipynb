```python
### Imports and Auxiliary Functions

import numpy as np
import matplotlib.pyplot as plt

%matplotlib nbagg
```

```python
### Defining the Log-likelihood function

def logL(alpha, cosines):
    m = np.size(cosines)
    value = - m * np.log(2) - m * np.log(1 + alpha/3) + np.sum( np.log( 1 + alpha * np.square( cosines ) ) )
    return value
```

```python
### Getting the cosine values

cosines = np.genfromtxt("list.txt", delimiter=',')
alpha = np.linspace(0, 20, 5000)
values = np.array([[logL(alpha[i], cosines)] for i in range(np.size(alpha))])
```

```python
fig = plt.figure()
plt.grid(True)
plt.xticks(np.arange(0, 25, 5))
plt.plot(alpha, values, 'k--', mfc='r')
plt.title('Maximum Likelihood Estimation')
plt.xlabel('alpha')
plt.ylabel('log(L(alpha))')
```

```python
pos = 0
maximum = values[pos]
for i in range(np.size(values)):
    if(values[i] > maximum):
        maximum = values[i]
        pos = i
max_alpha = alpha[pos]

print('The MLE of the parameter alpha is: ' + str(max_alpha))
```

```python
relative_values = values - np.max(values)
```

```python
pos = 0
maximum = relative_values[pos]
for i in range(np.size(relative_values)):
    if(relative_values[i] > maximum):
        maximum = relative_values[i]
        pos = i
temp = []
for j in range(np.size(relative_values)):
    if(np.abs(relative_values[pos] - relative_values[j]) < 0.5):
        temp.append(j)
uncertainty = np.array([alpha[temp[0]], alpha[temp[-1]]])
```

```python
fig = plt.figure()
plt.grid(True)
plt.xticks(np.arange(0, 25, 5))
plt.plot(alpha, relative_values, 'r--')
plt.plot(alpha, -.5 * np.ones_like(alpha), 'c')
plt.title('Relative Likelihood')
plt.xlabel('alpha')
plt.ylabel('log(L(alpha)) - max(log(L(alpha)))')
plt.plot(uncertainty[0] * np.ones_like(relative_values), relative_values, 'b')
plt.plot(uncertainty[1] * np.ones_like(relative_values), relative_values, 'b')
```

```python
print("Error bound on alpha is : [" + str(uncertainty[0]) + ', ' + str(uncertainty[1]) + ']')
```

```python
#Imports and Function definitions
%matplotlib nbagg 
import numpy as np
import matplotlib.pyplot as plt

def numder(x, h):
    return np.divide(np.exp(x+h) - np.exp(x), h)

def yprime(x):
    return np.exp(x)
```

```python
# Parameters
x = 1.
h = np.logspace(-1, -17, 1000000)
```

```python
# Calculating the Truncation error
trunc_error = np.divide(numder(x, h) - yprime(x), yprime(x))
```

```python
plt.semilogx(h, trunc_error)
```
